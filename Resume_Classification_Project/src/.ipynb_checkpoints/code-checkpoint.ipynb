#### Read the resume as pdf and parse it

```python
import PyPDF2
reader = PyPDF2.PdfReader('../data/Resume_latest.pdf')
resume_text = reader.pages[0].extract_text()
```

```python
import re
resume_text = resume_text.replace('\n',' ')
resume_text = resume_text.replace('%','percent ')
resume_text = re.sub('[^\w]',' ', resume_text)
```

```python
resume_text
```

#### Draw a Word Cloud on the resume input

```python
def print_word_cloud(text):
    word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text)
    plt.imshow(word_cloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
print_word_cloud(resume_text)
```

Getting the frequency of the words

```python
def word_counter(string):
    map = {}
    text_split = string.split(" ")
    for word in text_split:
        if word in map:
            map[word] += 1
        else:
            map[word] = 1
    keys = list(map.keys())
    values = list(map.values())
    sorted_value_index = np.argsort(values)
    sorted_dict = {keys[i]: values[i] for i in sorted_value_index}
    res =dict(reversed(list(sorted_dict.items())))
    return res
```

```python
word_counter(resume_text)
```

**As we can see, there are a lot of conjunctions and other words that occur too frequently, to process the text data, let us get rid of all the words except nouns**

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

from nltk import pos_tag
from nltk.tokenize import word_tokenize

def extract_nouns(words):
    """Extracts only the nouns from a list of words."""
    # Tokenize the words into individual tokens
    tokens = word_tokenize(words)
    # Use part-of-speech tagging to label each token with its part of speech
    tagged_tokens = pos_tag(tokens)
    # Only keep the tokens that are labeled as nouns (NN, NNS, NNPS, or NNP)
    nouns = [word for word, tag in tagged_tokens if tag.startswith('N')]
    # Join the remaining nouns back into a string
    return ' '.join(nouns)
```

```python
nouns_in_the_file = extract_nouns(resume_text)
```

```python
word_counter(nouns_in_the_file)
```

```python
print_word_cloud(nouns_in_the_file)
```
